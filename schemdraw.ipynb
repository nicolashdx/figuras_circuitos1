```python
import schemdraw
import schemdraw.elements as elm

from PIL import Image
import matplotlib.pyplot as plt
```

```python
with schemdraw.Drawing() as d:
    d.config(unit=4)
    V1 = elm.SourceV().label('5V')
    R1 = elm.Resistor().right().label('R1=1kΩ')
    elm.Dot()
    d.push()
    R2 = elm.Resistor().down()
    R2.label('R2=3,3kΩ', loc='bot', rotate=True)
    R2.label(['+','$v_o$','-'], loc='top')
    elm.Dot()
    d.pop()
    R3 = elm.Resistor().right().label('R3=2,2kΩ')
    V2 = elm.SourceV().down().reverse().label('10V', loc='bot')
    L2 = elm.Line().tox(V1.start)
    elm.LoopCurrent([R1,R2,L2,V1], pad=.8).label('$I_1$')
    elm.LoopCurrent([R3,V2,L2,R2], pad=.8).label('$I_2$')  
    #d.save("test.png")
```
